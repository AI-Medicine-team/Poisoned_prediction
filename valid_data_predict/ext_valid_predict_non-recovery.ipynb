## 2. pos weighted model prediction

```python
from catboost import CatBoostClassifier

# model_path = "/home/mailiyi/Poisoning_Prediction/ML/predict_non-recovery_best_model/catboost/depth_5_iterations_200_learning_rate_0.05/best_overall_model.cbm"
model_path = "/home/mailiyi/Poisoning_Prediction/ML/predict_non-recovery_best_model/catboost/depth_5_iterations_200_learning_rate_0.05/best_model_fold_5.cbm"

model = CatBoostClassifier()
model.load_model(model_path)
```

```python
import numpy as np
import pandas as pd
from sklearn import metrics

temp_df_x = dataX.copy()

for c in features_categorical:
    temp_df_x[c] = temp_df_x[c].astype(str).fillna("missing")

y_pred = model.predict_proba(temp_df_x)[:, 1]
y_true = dataY.values
# print(y_pred)

def bootstrap_metric_ci(y_true, y_pred, metric_fn, n_bootstrap=2000, seed=42):
    rng = np.random.RandomState(seed)
    scores = []

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    for _ in range(n_bootstrap):
        idx = rng.randint(0, len(y_true), len(y_true))
        if len(np.unique(y_true[idx])) < 2:
            continue
        scores.append(metric_fn(y_true[idx], y_pred[idx]))
    return (
        np.mean(scores),
        np.percentile(scores, 2.5),
        np.percentile(scores, 97.5)
    )

mean_auc, auc_lower, auc_upper = bootstrap_metric_ci(
    y_true, y_pred, metrics.roc_auc_score
)

mean_auprc, auprc_lower, auprc_upper = bootstrap_metric_ci(
    y_true, y_pred, metrics.average_precision_score
)

# print("AUROC:", mean_auc, auc_lower, auc_upper)
# print("AUPRC:", mean_auprc, auprc_lower, auprc_upper)
print(f"AUROC: Mean={mean_auc:.4f}, 95% CI=({auc_lower:.4f},{auc_upper:.4f})")
print(f"AUPRC: Mean={mean_auprc:.4f}, 95% CI=({auprc_lower:.4f},{auprc_upper:.4f})")

# y_true = dataY  
# y_pred = model.predict_proba(temp_df_x)[:, 1]

df_pred_results = pd.DataFrame({
    "y_test": y_true,
    "y_pred": y_pred
})
df_pred_results.to_csv(
    "/home/mailiyi/Poisoning_Prediction/valid_data_predict/predict_non-recovery_results.csv",
    index=False
)
df_pred_results
```
